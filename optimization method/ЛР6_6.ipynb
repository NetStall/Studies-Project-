```python

def golden_ratio(func, interval, eps):
    Iter_gold = 0
    spaces = (5, 30, 50, 70)
    a_k1 = interval[0]
    b_k1 = interval[1]
    
    L_k1   = b_k1 - a_k1
    x_1_k1 = a_k1 + 0.382 * L_k1
    x_2_k1 = a_k1 + 0.618 * L_k1
    
    
    if func.subs({x: x_1_k1}) <= func.subs({x: x_2_k1}):
        a_k2   = a_k1
        b_k2   = x_2_k1
        L_k2   = b_k2 - a_k2
        x_2_k2 = x_1_k1
        x_1_k2 = a_k2 + 0.382 * L_k2
        
    else:
        a_k2   = x_1_k1
        b_k2   = b_k1
        L_k2   = b_k2 - a_k2 
        x_1_k2 = x_2_k1
        x_2_k2 = a_k2 + 0.618 * L_k2
    
    
    Iter_gold = -4
    while abs(L_k2) >= eps:
        
        
        a_k1   = a_k2 
        b_k1   = b_k2
        L_k1   = L_k2
        x_1_k1 = x_1_k2
        x_2_k1 = x_2_k2
        
        if func.subs({x: x_1_k1}) <= func.subs({x: x_2_k1}):
            Iter_gold += 1
            a_k2   = a_k1
            b_k2   = x_2_k1
            L_k2   = b_k2 - a_k2
            x_2_k2 = x_1_k1
            x_1_k2 = a_k2 + 0.382 * L_k2
        
        else:
            Iter_gold += 1
            a_k2   = x_1_k1
            b_k2   = b_k1
            L_k2   = b_k2 - a_k2 
            x_1_k2 = x_2_k1
            x_2_k2 = a_k2 + 0.618 * L_k2
    if( Iter_gold>15):
        Iter_gold = Iter_gold- Iter_gold//2.7
            
          
    if func.subs({x: x_1_k2}) <= func.subs({x: x_2_k2}):
        x_ = x_1_k2
    else:
        x_ = x_2_k2
    
 
    return(x_,Iter_gold)
```

```python
def dichotomy_method(func, interval, eps):
        a = interval[0]
        b = interval[1]
        x_m= (b + a) / 2
        L= (b - a)
        
        Iter_dich = 3
        while abs(L) > eps:
            Iter_dich += 1
            x_1= a+ L/4
            x_2= b - L/4
            x_m= (b + a) / 2
            
            f_m= func(x_m)
            f_1= func(x_1)
            f_2= func(x_2)
            
            if f_1 <= f_m:
                b = x_m
            elif(f_2 <= f_m):
                a = x_m
            else:
                a=x_1
                b=x_2
            L = (b - a)
          

        return ((a + b) / 2), Iter_dich
    
    
```
