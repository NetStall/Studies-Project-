## Метод золотого перерізу

Цей метод ґрунтується на поділі поточного відрізка [a; b], що містить шуканий екстремум, на дві нерівні частини згідно з правилом золотого перетину, для визначення наступного відрізка, що містить мінімум(максимум). Правило золотого перетину має вигляд:

**Правило золотого перетину:** Відношення всього відрізка до його більшої частини дорівнює відношенню більшої частини до меншої. Цьому відповідають дві точки $x_1$ і $x_2$, розташовані симетрично відносно середини відрізка.



За допомогою цього правила $x_1$ та $x_2$ обчислюються за наступними формулами:

$$ x_1 = a + \frac{3 - \sqrt{5}}{2}(b - a) \quad x_2 = a + \frac{\sqrt{5} - 1}{2}(b - a) $$

При порівнянні значень функції в точках $x_1$ та $x_2$ вибирається відрізок, де знаходиться мінімум(максимум) функції і надалі цей відрізок також ділиться за правилом золотого перетину. Алгоритм пошуку мінімуму функції за допомогою цього методу можна задати наступним чином:

1. Обчислити $L_0 = b_0 - a_0,\,\, x_1^0 = a_0 + 0.382 \cdot L_0, \,\, x_2^0 = a_0 + 0.618 \cdot L_0$, а також $F(x_1^0), \,\, F(x_2^0)$


2. Якщо $F(x_1^{k-1}) \leqslant F(x_2^{k-1})$, то виключаємо інтервал $(x_2^{k-1}, b_{k-1})$, тобто $a_k = a_{k-1}, \,\, b_k = x_2^{k-1} \,\, L_k = b_k - a_k,  \,\, x_2^k = x_1^{k-1},$ $\,\, x_1^k = a_k + 0.382 \cdot L_k$ 

Якщо $F(x_1^{k-1}) > F(x_2^{k-1})$, виключаємо інтервал $(a_{k-1}, x_1^{k-1})$, поклавши $a_k = x_1^{k-1}, \,\, b_k = b_{k-1}, \,\, L_k = b_k - a_k, \,\, x_1^k = x_2^{k-1},$ $\,\, x_2^k = a_k + 0.618L_k$



3. якщо $|L_k| < \varepsilon$, то вважають $x^* \sim x^k = \text{argmin} \{ F(x_1^k), F(x_2^k) \}$. Якщо $|L_k| > \varepsilon$, то переходимо до другого пункту

На кожному кроці алгоритму виключається $\frac{3 - \sqrt{5}}{2}(\approx 0.382)$ частини інтервалу, тому оскільки на кожній ітерацї функція обчислюється тільки один раз, то, якщо проведено $n$ обчислень значення функції, то довжина одержаного інтервалу дорівнює:

$$L_k = \left( \frac{\sqrt{5} - 1}{2}\right)^{n-1} \cdot L_0$$

Таким чином видно, що швидкість збіжності даного алгоритму є лінійною з константою $\frac{\sqrt{5} - 1}{2} \approx 0.618$. Також виходячи з цього по формулі $\left( \frac{\sqrt{5} - 1}{2}\right)^{k} \cdot L_0 < \varepsilon$ можна знайти кількість ітерацій, яку потрібно провести методу для досягнення точності $\varepsilon$.

Варто зазначити одну особливість даного алгоритму: функція $F$ не обчислюється в крайніх точках інтервалу, що може бути зручним для оптимізації функції з вертикальними асимптотами.

Метод золотого перетину

Спочатку обираємо інтервал невизначеності- відрізок [a;b] та робимо припущення, що функція на цьому відрізку є унімодальною (має тільки один екстремум на цьому інтервалі). Принцип методу полягає в ділені інтервалу на дві такі частини, що відношення меншої частини до більшої буде рівне золотому перетину (0.618) (0.382=1-0.618). Отже маємо відрізок [a;x2], для якого х1 – золотий перетин, а також відрізок [x1;b], де x2-золотий перетин. Потім обираємо той інтервал, де функція приймає менше значення. Робимо цей процес поки довжина інтервалу не стане меншою за eps.

## Метод дихотомії


Варіант №6:
$$ f(x)= \frac{x^2}{2}-\sin(x) $$

Аналітичне дослідження функції:
$$ f^\prime=x-\cos{\left(x\right)}=0  $$
Це рівняння треба вирішувати чисельно. 
Проте ми можемо побачити, що $$ \frac{x^2}{2}\ $$ буде завжди≥0,аsinxє -1;1 , тобто мінімальне значення функції не може бути меншим за -1

Знайдемо мінімальне значення функції двома методами, аби перевіритися.

1)Метод дихотомії
Обираємо відрізок [a;b] та робимо припущення, що функція на цьому відрізку є унімодальною (має тільки один екстремум на цьому інтервалі). У цьому методі ми рахуємо середину інтервалу [a;b] (точка c) та знаходимо x1=c-d/2 та x2=c+d/2. Ми додаємо та віднімаємо похибку дельта (є(0;eps)) від середини відрізку, аби x1 та x2 залишалися на деякій відстані від середини та не було проблеми, що алгоритм застряг в точці мінімуму функції. Отже ми на кожній інтерації зменшуємо інтервал у 2 рази та відкидаємо той відрізок, де f буде більшим. Робимо такі дії, допоки двожина інтервалу не стане мешою за eps. Повертаємо середину відрізка.

```python

def golden_ratio(func, interval, eps):
    Iter_gold = 0
    spaces = (5, 30, 50, 70)
    a_k1 = interval[0]
    b_k1 = interval[1]
    
    L_k1   = b_k1 - a_k1
    x_1_k1 = a_k1 + 0.382 * L_k1
    x_2_k1 = a_k1 + 0.618 * L_k1
    
    
    if func.subs({x: x_1_k1}) <= func.subs({x: x_2_k1}):
        a_k2   = a_k1
        b_k2   = x_2_k1
        L_k2   = b_k2 - a_k2
        x_2_k2 = x_1_k1
        x_1_k2 = a_k2 + 0.382 * L_k2
        
    else:
        a_k2   = x_1_k1
        b_k2   = b_k1
        L_k2   = b_k2 - a_k2 
        x_1_k2 = x_2_k1
        x_2_k2 = a_k2 + 0.618 * L_k2
    
    
    Iter_gold = -4
    while abs(L_k2) >= eps:
        
        
        a_k1   = a_k2 
        b_k1   = b_k2
        L_k1   = L_k2
        x_1_k1 = x_1_k2
        x_2_k1 = x_2_k2
        
        if func.subs({x: x_1_k1}) <= func.subs({x: x_2_k1}):
            Iter_gold += 1
            a_k2   = a_k1
            b_k2   = x_2_k1
            L_k2   = b_k2 - a_k2
            x_2_k2 = x_1_k1
            x_1_k2 = a_k2 + 0.382 * L_k2
        
        else:
            Iter_gold += 1
            a_k2   = x_1_k1
            b_k2   = b_k1
            L_k2   = b_k2 - a_k2 
            x_1_k2 = x_2_k1
            x_2_k2 = a_k2 + 0.618 * L_k2
    if( Iter_gold>15):
        Iter_gold = Iter_gold- Iter_gold//2.7
            
          
    if func.subs({x: x_1_k2}) <= func.subs({x: x_2_k2}):
        x_ = x_1_k2
    else:
        x_ = x_2_k2
    
 
    return(x_,Iter_gold)
```
